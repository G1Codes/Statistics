### Using statsmodel

```python
# using statsmodel

import statsmodels.api as sm
import matplotlib.pyplot as plt

# Create a QQ plot of the two sets of data
fig = sm.qqplot(df['sepal_length'], line='s')

# Add a title and labels to the plot
plt.title('QQ Plot')
plt.xlabel('Theoretical Quantiles')
plt.ylabel('Sample Quantiles')

# Show the plot
plt.show()
```

```python
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt

# Generate normally distributed data
data = np.random.normal(0, 1, 1000)

# Different Q-Q plots with different lines
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

sm.qqplot(data, line='45', ax=axes[0, 0])
axes[0, 0].set_title("line='45' (45-degree diagonal)")

sm.qqplot(data, line='s', ax=axes[0, 1])
axes[0, 1].set_title("line='s' (Standardized)")

sm.qqplot(data, line='r', ax=axes[1, 0])
axes[1, 0].set_title("line='r' (Regression)")

sm.qqplot(data, line='q', ax=axes[1, 1])
axes[1, 1].set_title("line='q' (Quartile-based)")

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import scipy.stats as stats
import statsmodels.api as sm
import matplotlib.pyplot as plt

# Generate a set of random data
x = np.random.uniform(low=0, high=1, size=1000)
```

```python
plt.hist(x)
```

```python
# Fit a uniform distribution to the data
params = stats.uniform.fit(x)
dist = stats.uniform(loc=params[0], scale=params[1])
```

```python
params
```

```python
dist
```

```python
# Create a QQ plot of the data using the uniform distribution
fig = sm.qqplot(x, dist=dist, line='45')

# Add a title and labels to the plot
plt.title('QQ Plot of Uniform Distribution with Uniform Fit')
plt.xlabel('Theoretical Quantiles')
plt.ylabel('Sample Quantiles')

# Show the plot
plt.show()
```

## Log Normal Distribution

```python
import numpy as np  
import matplotlib.pyplot as plt  
import seaborn as sns  

# Parameters for the underlying Normal Distribution  
mu = 1.0    # Mean of ln(X)  
sigma = 0.5 # Std dev of ln(X)  

# Generate Log-Normal data  
data = np.random.lognormal(mean=mu, sigma=sigma, size=1000)  
```

```python
plt.figure(figsize=(12, 5))  

# Plot Log-Normal Data  
plt.subplot(1, 2, 1)  
sns.histplot(data, kde=True, color='blue')  
plt.title('Log-Normal Distribution')  

# Plot Normal Data (Log-Transformed)  
plt.subplot(1, 2, 2)  
sns.histplot(np.log(data), kde=True, color='green')  
plt.title('Normal Distribution (ln(X))')  

plt.tight_layout()  
plt.show()  
```
